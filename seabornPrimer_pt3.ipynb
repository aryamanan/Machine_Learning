## `sns.boxplot()`: Bar Plot

```python
df=sns.load_dataset("penguins")
print(df.head())
```

```python
sns.barplot(x=df["island"],y=df["bill_length_mm"])
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex")
# categorise by "sex" using hue
plt.show()
```

### Change order of the bar graph

```python
ord=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",order=ord,hue_order=["Female","Male"])
# order changes X-label object order and colour order
plt.show()
```

#### when changing orientation, remember that the orientation needs a numerial value, it won't work with categorical values

> for example, if orientation is horizontal (h), then x needs to be numerical, if orientation is vertical then y needs to be numerical

```python
ord=["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_length_mm",y="island",data=df,hue="sex",order=ord,hue_order=["Female","Male"],orient="h")
# change bar-axis to horizontal using orient
plt.show()
```

- orient="h": This parameter determines the orientation of the bars in the plot.

- saturation=10: Saturation refers to the intensity or purity of colors.
 `saturation value of 10 suggests that the colors used for the bars will be highly intense or vivid`

- errcolor="g": The errcolor parameter specifies the color to be used for error bars, if any. `g means green`

- errwidth=2: The errwidth parameter sets the thickness or width of the error bars.
`A value of 2 implies that the error bars will be relatively thicker`

- capsize=0.2: The capsize parameter determines the size of the caps or edges at the ends of the error bars.
`A value of 0.2 indicates that the caps will have a length or size equal to 20% of the total width of the error bar`

- dodged=True, each category or group in the data is represented by separate bars placed adjacent to each other, making it easier to compare values across different categories.

- alpha: lighten the colour of the graph



```python
sns.set(style="darkgrid")
ord=["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_length_mm",y="island",data=df,hue="sex",\
            order=ord,hue_order=["Female","Male"],orient="h",\
            saturation=10, errcolor="g",errwidth=2,\
            capsize=0.2,alpha=0.7)
plt.show()
```
