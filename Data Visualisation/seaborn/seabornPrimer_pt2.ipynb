# Seaborn Primer pt2

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("penguins")
```

## `sns.histplot` and `sns.displot` : Histograms
> **Difference in Bar Plots and Histograms:** Bar Plots are for discrete/categorical data with related continuous values, while Histograms are used for continuous data related to continuous values

```python
# plotting a histogram

sns.histplot(df["bill_depth_mm"])
plt.show()
```

```python
# custom no of bins
sns.histplot(df["bill_depth_mm"],bins=35)
plt.show()
```

```python
# custom class interval bins
sns.histplot(df["bill_depth_mm"],bins=[12,15,17,19])
```

## `sns.kdeplot`: Kernel Density Estimation (KDE)
- non-parametric method to estimate distribution of a dataset
- It outputs the PDF (probability distribution function) using kernels (like Gaussian, Triangular, Epanechnikov, Uniform, Quartic etc)
- What these Kernels do is they calculate kernel function for each point Xi in the original points in dataset and sum it all up and normalises it all by no of points in dataset)
- the resulting function represents the PDF of the function,a
- PDF of a dataset gives probabilities of having an entry in dataset at a given point on number-line

```python
sns.kdeplot(df["bill_depth_mm"])
plt.show()
```

```python
# custom class interval bins
sns.histplot(df["bill_depth_mm"],bins=15,kde=True)
plt.show()
```

```python
sns.displot(df["bill_depth_mm"],bins=15,kde=True,rug=True)
# rug shows the KDE values on numberline
plt.show()
```

```python
sns.displot(df["bill_depth_mm"], bins=15, kde=True, rug=True,color='r')
# change colour to red
plt.show()
```

### Logarithmic Scale in Histograms
- Instead of a linear scale where each increment represents a constant difference, a logarithmic scale increases exponentially.
 - The logarithmic scale compresses the range of values, emphasizing smaller values and reducing the visual impact of larger values.

```python
sns.displot(df["bill_depth_mm"], bins=15, kde=True, rug=True,color='g',log_scale=True),
# changes histogram from linear to log-scale
plt.show()
```
