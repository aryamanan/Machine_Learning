# Basic Plots using `matplotlib` 

```python
import numpy as np
import matplotlib.pyplot as plt

# OR

from matplotlib import pyplot as plt
```

```python
# Lets make some sample data to plot by ourselves


# independent varible (x)
age=np.abs(np.random.normal(30,30,65)) # (mean,std-deviation,no-of-elements)
age_x=np.sort(age)
# dependent variable (y)
salary=np.abs(np.random.exponential(45000,65)) 
salary_y=np.sort(salary)
```

### Making a simple graph with common commands

```python
# Common Commands

plt.plot(age_x,salary_y,label='Web Dev') # plot the (x,y) co-ordinate graph, and label it

# Legend is a graphical element that describes the meaning of the various symbols or colors used in a graph.
plt.legend()

plt.xlabel('Age') # give a label to x-axis
plt.ylabel('Median Salary') # give a label to y-axis
plt.title('CSE Passouts Median Salary vs Their Age') # give a title to the whole graph
plt.show() # show the graph
```

```python
# now let's make another one

salary2=np.abs(np.random.exponential(85000,65))
salary2_y=np.sort(salary2)

plt.plot(age_x,salary2_y,label='Data Scientist')
plt.xlabel('Age')
plt.ylabel('Median Salary')
plt.title('CSE Passouts Median Salary vs Their Age')
plt.legend()
plt.show()
```

# Matplotlib Subplots

- #### When you create a figure and one or more subplots using Matplotlib's pyplot.subplots() function, it returns two objects: a Figure object and one or more Axes objects.


## Figure

- In Matplotlib, a figure represents a single window or page in a visualization. 
- The Figure object represents the entire figure, including any subplots and their layout.
- A figure can contain one or more subplots, which are also called axes. 

## Axes

- The Axes object represents a single subplot or axis within the figure. 
- An axis is an area in the figure where the data is plotted.


```python
# We'll make the matplotlib graph using figure,axis objects this time


# Create the figure and axes objects
fig,ax=plt.subplots() 

# Plot the first graph
ax.plot(age_x,salary_y,color='blue',label='Web Developer')

# Plot the second graph on the same independent axes/subplot
ax.plot(age_x,salary2_y,color='red',label='Data Scientist')

# Add a legend 
ax.legend()

# Add axis labels and a title
ax.set_xlabel('Median Salary')
ax.set_ylabel('Age')
ax.set_title('CSE Passouts Median Salary vs Their Age')

# Display the graph
plt.show()
```
