# Seaborn Primer pt 1

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var=[1,2,3,4,5,6,7]
var1=[2,3,4,5,3,9,8]
```

```python
plt.figure(1,figsize=(20,5))
plt.plot(var,var1)
plt.show()
```

## `seaborn` can only work with `pandas` dataframes only

```python
import pandas as pd
x=pd.DataFrame({"var":var, "var1":var1}) # setting var and var1 as pandas DataFrame columns
print(x.head())
```

```python
plt.figure(1,figsize=(20,5))
sns.lineplot(x="var",y="var1",data=x)
# x and y here will be plt-labels of the seaborn plot
# IMPORTANT: x and y need to be column names from a pandas dataframe, and data=dataframe
plt.show()
```

```python
X=sns.load_dataset('mpg').dropna()
print(X.head())
```

```python
# Line-Plot: A simple cartesian co-ordinate graph with x and y
sns.set_style('darkgrid') # can be 'whitegrid' as well
plt.figure(1,figsize=(20,5))
sns.lineplot(x='cylinders',y='displacement',data=X)
plt.show()
```

```python
# We can add hue to group data wrt a categorical variable feature column in the dataframe
# this allows us to compare some broad classes within the dataset
plt.figure(1,figsize=(20,5))
sns.lineplot(x='cylinders',y='displacement',data=X,hue='origin')
plt.show()
```

```python
# We can add 'style' wrt categorical variable too to make further distinctions
plt.figure(1,figsize=(20,5))
sns.lineplot(x='cylinders',y='displacement',data=X,hue='origin',style='model_year')
plt.show()
```

```python
# This is too messy, let's broaden the year category by adding another column in our dataframe
X['decade'] = pd.Series(dtype='object')

# Define conditions and assign values to 'decade' based on 'model_year'
X.loc[(X['model_year'] >= 70) & (X['model_year'] <= 79), 'decade'] = '70s'
X.loc[(X['model_year'] >= 80) & (X['model_year'] <= 89), 'decade'] = '80s'
```

```python
plt.figure(1,figsize=(20,5))
sns.lineplot(x='cylinders',y='displacement',data=X,hue='origin',style='decade')
plt.show()
```

```python
# In a lineplot, the faded region typically represents the confidence interval around the estimated line or curve.
# The darker or more saturated portion of the line represents the estimated values, while the faded region around it represents the uncertainty or variability.

# But I wanna get rid of the confidence interval faded region here, I don't need it
plt.figure(1,figsize=(20,5))
sns.lineplot(x='cylinders',y='displacement',data=X,hue='origin',style='decade',ci=None)
plt.show()
```

```python
# palette --> add more distinctive colours colours
# marker --> add marker for further distinction
plt.figure(1,figsize=(20,5))
sns.lineplot(x='cylinders',y='displacement',data=X,hue='origin',style='decade',
             errorbar=None, palette="Accent",markers=["o",">"],dashes=True,legend=True)
plt.show()
```
